```python
# plot a heatmap
plt.figure(figsize=(15,12))
sns.heatmap(corr_matrix)
```

```python
#select the upper triangle of corr matrix
upper=corr_matrix.where(np.triu(np.ones(corr_matrix.shape),k=1).astype(bool))
upper
```

```python
# plot a heatmap
plt.figure(figsize=(15,12))
sns.heatmap(upper)
```

```python
# finding the col having corr coff. more than 0.9 or less than -0.9
```

```python
neg_correleted=[column for column in upper.columns
               if any(upper[column]<=-.9)]
pos_correleted=[column for column in upper.columns
               if any(upper[column]>=.9)]
print(neg_correleted)
print("***************" *5)
print(pos_correleted)
```

```python
total_correleted=neg_correleted+pos_correleted
```

```python
print(total_correleted)
```

```python
df=data1.drop(total_correleted,axis=1)
```

```python
df.shape
```

```python
# VIF approach
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif=pd.DataFrame()
vif["variables"]=df.columns
vif["VIF"]=[variance_inflation_factor(df.values,i)
           for i in range(df.shape[1])]
vif
```

```python
pd.set_option("display.max_rows",600)
```

```python
vif
```

```python
high_vif=vif[vif["VIF"]>10].sort_values("VIF")
high_vif
```

```python
high_vif.shape
```

```python
# we must add a constant when we calculate the VIF
import statsmodels.api as sm
df1=sm.add_constant(df)
```

```python
vif1=pd.DataFrame()
vif1["variables"]=df1.columns
vif1["VIF"]=[variance_inflation_factor(df1.values,i)
           for i in range(df1.shape[1])]
vif1
```

```python
high_vif_1=vif1[vif1["VIF"]>10].sort_values("VIF")
high_vif_1
```

```python
high_vif_1.shape
```

```python
# calculate the accuracy when drop the dimensions
```

```python
df2=df1.drop(high_vif_1["variables"],axis=1)
```

```python
df2.shape
```

```python
df2.head()
```

```python
#Recurrsive Feature elmination approach
```

```python
# dependent and independent variable
```

```python
X=df2.drop(["Pass/Fail"],axis=1)
Y=df2["Pass/Fail"]
```

```python
from sklearn.feature_selection import SelectFromModel,RFECV
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,random_state=101,test_size=0.2,stratify=Y)
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
sfm=SelectFromModel(RandomForestClassifier(n_estimators=100,random_state=True,n_jobs=-1))
sfm.fit(X_train,Y_train)
```

```python
sfm.get_support()
# features which paas the benchmark will show as True otherwise False
```

```python
print("benchmark:mean of feature importance:  ", np.mean(sfm.estimator_.feature_importances_))
```

```python
sfm.estimator_.feature_importances_
```

```python
feature_selected_sfm=X_train.columns[sfm.get_support()]
```
